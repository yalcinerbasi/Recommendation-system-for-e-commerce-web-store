```python
##Python Dependecies
import json
import numpy as np
import pandas as pd
import os
import gc
```

```python
##Change directory to read json data
os.chdir("./data/data")
```

```python
##Read event data and convert to dataframe
event = json.load(open("events.json",encoding="utf8"))
df_event = pd.json_normalize(event["events"])
df_event
```

```python
##Read product data and convert to dataframe
meta = json.load(open("meta.json",encoding="utf8"))
df_meta = pd.json_normalize(meta["meta"])
df_meta
```

```python
##Creating a matrix by adding 1 to the pivot table for each product added to the cart
columns = df_meta["productid"].values
rows = list(dict.fromkeys(df_event["sessionid"].tolist()))
df_matrix = pd.DataFrame(0,columns=columns,index=rows)
for i in range(len(df_event)):
    df_matrix.loc[df_event.iloc[i,1],df_event.iloc[i,4]] += 1
df_matrix.head()
```

```python
##Normalize matrix to a common scale
df_matrix_norm = df_matrix / np.sqrt(np.square(df_matrix).sum(axis=0))
```

```python
#Compute Cosine Similarities by adding with Vectors
df_cosine_sim = df_matrix_norm.transpose().dot(df_matrix_norm)
df_cosine_sim.head()
```

```python
##Create a placeholder products for closes neighbours to an products
def close_neighbours(df):
    products_neighbours =  pd.DataFrame(index=df.columns,columns=range(0,11))
    for i in range(0,len(df)):
        products_neighbours.iloc[i,:11] = df.iloc[0:,i].sort_values(ascending=False)[0:11].index
    return products_neighbours

##Create similarity matrix with scores
def similarity_score(df):
    score =  pd.DataFrame(index=df.columns,columns=range(0,11))
    for i in range(0,len(df)):
        score.iloc[i,:11] = df.iloc[0:,i].sort_values(ascending=False)[0:11]
    return score
```

```python
##Calculate result and save it 
os.chdir("../../result_data")
close_neighbours = close_neighbours(df_cosine_sim)
close_neighbours.to_csv("Cosine_Sim_Name.csv")
close_neighbours
```

```python
similarity_score = similarity_score(df_cosine_sim)
similarity_score.to_csv("Cosine_Sim_Score.csv")
similarity_score
```

```python
## Empty Memory
del df_matrix_norm, df_cosine_sim
gc.collect()
```
